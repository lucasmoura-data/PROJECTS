Casos aonde só quero determinado resultado.

```python
livros = ["1984", "Dom Casmurro", "O Pequeno Príncipe", "O Hobbit", "Orgulho e Preconceito"]

for livro in livros:
    if livro == "O Hobbit":
        print(f"Livro encontrado {livro}")
        break # assim que encontrado o procurado ele para, ou seja, só traz o resultado encontrado. Do contrário, resultado vazio
```

Simulador de estoque de vendas  

```python
estoques = 5

# while pode ser traduzido com o nome de: enquanto.

while estoques > 0: # se não houvesse o estoques -=1, o valor se repetiria por infinito, pois 5 sempre será maior que 0
    estoques -= 1
    print(f"Venda realizada! Estoque restante: {estoques}")

print("Produto Esgotado")
```

```python
contador = range(10,0,-1)

for cont in contador:
    if cont % 2 == 0:
        print(f"Faltam apenas {cont} segudos - Não perca essa oportunidade")
    elif cont % 2 == 1:
        print(f"A contagem continua: {cont} segundos restantes")

print("Aproveite a promoção agora")
```

```python
livros = [
    {"nome": "1984", "estoque": 5},
    {"nome": "Dom Casmurro", "estoque": 0},
    {"nome": "O Pequeno Príncipe", "estoque": 3},
    {"nome": "O Hobbit", "estoque": 0},
    {"nome": "Orgulho e Preconceito", "estoque": 2}
]


for livro in livros:
    if livro["estoque"] != 0:
        print(f"Livro disponível: {livro["nome"]}")
```

```python
livros = [
    {"nome": "1984", "nota1": 5, "nota2":10},
    {"nome": "Dom Casmurro", "nota1": 0, "nota2":10},
    {"nome": "O Pequeno Príncipe", "nota1": 3, "nota2":10},
    {"nome": "O Hobbit", "nota1": 0, "nota2":10},
    {"nome": "Orgulho e Preconceito", "nota1": 2, "nota2":10}
]

for livro in livros:
    media = livro["nota1"] + livro["nota2"] / 2
    print(f"A media do livro {livro["nome"]} é de: {media}")
```
